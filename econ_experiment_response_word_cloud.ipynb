This code generates wordclouds for timebased responses and milestone responses. 

```python
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import matplotlib.pyplot as plt
```

-------------------------------------------------------------------------------------------------------------------------------------------

First, we look at the timebased responses. 

```python
# read the excel file
df = pd.read_excel('241124_survey.xlsx')

df_timebased = df[(df['chosen_feedback'] == 'Timebased feedback')]
vote_feedback_explain_timebased = '\n'.join(df_timebased['vote_feedback_explain'].astype(str))
```

```python
stopwords = set(STOPWORDS)
#stopwords.update(["will", "see", "flavors"])

wc = WordCloud(
    background_color = 'white',
    stopwords        = stopwords, 
    height           = 500,
    width            = 500,
)

wc.generate(vote_feedback_explain_timebased)
```

```python
plt.plot()
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

-------------------------------------------------------------------------------------------------------------------------------------------

Next, we look at the milestone responses. 

```python
df_milestone = df[(df['chosen_feedback'] == 'Milestone feedback')]
vote_feedback_explain_milestone = '\n'.join(df_milestone['vote_feedback_explain'].astype(str))
```
